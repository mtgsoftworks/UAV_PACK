```python
import numpy as np
from ultralytics import YOLO 
import cv2
import os
from IPython.display import Video
import matplotlib.pyplot as plt
import shutil
import keyboard
```

```python
#Model buradan verilmelidir.
model = YOLO("last.pt")
```

```python

cap = cv2.VideoCapture("iha_video.mp4")

i = 0
p = 0
img_way = []
label_way = []
while True:
    
    ret, frame = cap.read()
    images_path = ".\\images\\frame" + str(int(p)) + ".jpg"

    i += 1
    if i % 60 != 0:
        continue
    if ret:
        p += 1 
        frame_resized = cv2.resize(frame, (416, 416))
        cv2.imwrite(images_path, frame_resized)
        img_way.append(images_path)
    else:
        break
cap.release()
label_way = img_way.copy()

p = 0
print(img_way)
while True:

    length_image = len(img_way)
    
    positive_images_path = ".\\positive\\images\\frame" + str(int(p)) + ".jpg"
    negative_images_path = ".\\negative\\images\\frame" + str(int(p)) + ".jpg"
    mixed_images_path = ".\\mixed\\images\\frame" + str(int(p)) + ".jpg"

    positive_labels_path = ".\\positive\\labels\\frame" + str(int(p)) + ".txt"
    negative_labels_path = ".\\negative\\labels\\frame" + str(int(p)) + ".txt"
    mixed_labels_path = ".\\mixed\\labels\\frame" + str(int(p)) + ".txt"

    print(img_way[p] + ": " + str(p))
    frame = cv2.imread(img_way[p])
    

    frame_resized = cv2.resize(frame, (416, 416))
    result_predict = model(frame)
    plot = result_predict[0].plot()
        
    obj = ""
    for result in result_predict:
        if len(result.boxes.xyxy) != 0:
            xyxy = result.boxes.xyxy
            xyxy = xyxy[0].tolist()
            x1, y1, x2, y2 = xyxy

            obj += "0 " + str(x1/416) + " " + str(y1/416) + " " + str(x2/416) + " " + str(y1/416) + " " + str(x1/416) + " " + str(y2/416) + " " + str(x2/416) + " " + str(y2/416) + "\n"

    cv2.imshow("img", plot)

    key = cv2.waitKey(0)   

    #Tuş değişikliği için buradan değiştirilebilir.

    if key == ord("a") and p != length_image -1:
        shutil.move(img_way[p], ".\\positive\\images\\frame" + str(int(p)) + ".jpg")
        
        img_way[p] = ".\\positive\\images\\frame" + str(int(p)) + ".jpg"
        label_way[p] = positive_labels_path

        with open(positive_labels_path, "w") as dosya:
            dosya.write(obj)

        p += 1
    elif key == ord("s") and p != length_image -1:
        shutil.move(img_way[p],  ".\\negative\\images\\frame" + str(int(p)) + ".jpg")
        
        img_way[p] = ".\\negative\\images\\frame" + str(int(p)) + ".jpg"
        label_way[p] = negative_labels_path

        with open(negative_labels_path, "w") as dosya:
            dosya.write(obj)
        p += 1
    elif key == ord("d") and p != length_image -1:
        shutil.move(img_way[p], ".\\mixed\\images\\frame" + str(int(p)) + ".jpg")
        
        img_way[p] = ".\\mixed\\images\\frame" + str(int(p)) + ".jpg"
        label_way[p] = mixed_labels_path

        with open(mixed_labels_path, "w") as dosya:
            dosya.write(obj)

        p += 1
    elif keyboard.is_pressed('left') and p != 0:
        p = p - 1

    elif keyboard.is_pressed('right') and p != length_image -1:
        p = p + 1

    elif key == ord("q"):
        break
    

cv2.destroyAllWindows()
```
